```python
# Plotting
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline


# Basic imports
import numpy as np
from copy import deepcopy
import pandas as pd
from tqdm import tqdm

#JAX
import jax
import jax.numpy as jnp
from jax.config import config
config.update("jax_enable_x64", True)
config.update("jax_debug_nans", True)

#My GRF
from jaxtronomy.GaussianRandomField.PowerBox_jaxified import get_jaxified_GRF
from jaxtronomy.GaussianRandomField.PowerBox_jaxified import get_k_grid_dx


#GRF generator
import powerbox as pbox

# Universal font size
FS = 18
```

```python
npix=100
pix_scl = 0.08  # arcsec / pixel

seed=1
```

# Power slope

```python
A=1e-5
beta=7.
seed=1

GRF=get_jaxified_GRF([A,beta],seed,npix,pix_scl)
GRF_pbox=pbox.PowerBox(N=npix, dim=2, pk=(lambda k: A * k**(-beta)), boxlength=2 * np.pi * npix * pix_scl, seed=seed,vol_normalised_power=False).delta_x()


fig,ax=plt.subplots(1,3,figsize=(15,7))

im0=ax[0].imshow(GRF)
im1=ax[1].imshow(GRF_pbox)
im2=ax[2].imshow(GRF-GRF_pbox,cmap='seismic',norm=mpl.colors.TwoSlopeNorm(0))

names=['My code','Powerbox','resid']
for i,im in enumerate([im0,im1,im2]):
  ax[i].set_title(names[i])
  ax[i].axis('off')
  fig.colorbar(im,ax=ax[i],orientation='horizontal')

fig.tight_layout()
fig.suptitle('GRF: logA={:.0f}, '.format(np.log10(A))+r'$\beta=$'+'{:.0f}'.format(beta),y=1.1,fontsize=FS)
plt.show()
```

# Amplitude

```python
A=1e-5
beta=7.
seed=1

GRF=get_jaxified_GRF([A,beta],seed,npix,pix_scl)
GRF_pbox=pbox.PowerBox(N=npix, dim=2, pk=(lambda k: A * k**(-beta)), boxlength=2 * np.pi * npix * pix_scl, seed=seed,vol_normalised_power=False).delta_x()


print('GRF variance {:.2f}'.format(GRF.std()**2))
print('GRF_pbox variance {:.2f}'.format(GRF_pbox.std()**2))
```

```python
k_grid,dx=get_k_grid_dx(npix,pix_scl)
```

```python
A=1e-5
beta=7.

GRF=get_jaxified_GRF([A,beta],seed,npix,pix_scl)
print('GRF variance {:.2f}'.format(GRF.std()**2))

Power_spectrum=A*np.power(k_grid,-beta)
Power_spectrum[npix//2,npix//2]=0
print('Predicted variance',Power_spectrum.sum())

print('Corected variance',Power_spectrum.sum()/GRF.std()**2)
```

```python
Betas=np.linspace(2,3,50)
GRF_vars=[np.std(get_jaxified_GRF([A,beta],seed,npix,pix_scl))**2 for beta in Betas]
Power_vars=np.zeros_like(Betas)
for i,beta in enumerate(Betas):
  Power_spectrum=A*np.power(k_grid,-beta-2)
  Power_spectrum[npix//2,npix//2]=0
  Power_vars[i]=Power_spectrum.sum()
```

```python
Power_vars/GRF_vars
```

```python
plt.plot(Power_vars,GRF_vars)
```
